```python
#Sum fuel_cost_per_unit_burned for fuel type Coal in year 1994
coal_burned_in_1994 = 14984.572
#Sum fuel_cost_per_unit_burned for fuel type Coal in year 1998
coal_burned_in_1998 = 11902.597
```

```python
percent_change_coal_burned = ((coal_burned_in_1998 - coal_burned_in_1994) / coal_burned_in_1994) * 100
```

```python
print(percent_change_coal_burned)
```

```python
#round up to the nearest whole number
print(round(percent_change_coal_burned, 0))
```

```python
# Print Out
print("For fuel type coal, the percentage change in the fuel cost per unit burned in 1998 compared to 1994 is:",(round(percent_change_coal_burned, 0)) ,"%")
```
